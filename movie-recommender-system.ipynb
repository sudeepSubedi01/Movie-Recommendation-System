# Keeping only required features

```python
# required columns: genres, id, keywords, title, overview, cast, crew
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.sample()
```

# Preparing Genres

```python
print(type(movies.iloc[0]['genres']))
# The list of genres is a string that needs to be converted to a python list to access its elements
import ast
ast.literal_eval(movies.iloc[0]['genres'])
```

```python
import ast
def convert_str_to_list(obj):
  new_list = []
  for i in ast.literal_eval(obj):
      new_list.append(i['name'])
  return new_list
```

```python
# convert_to_list(movies.iloc[0]['genres'])
movies['genres'] = movies['genres'].apply(convert_str_to_list)
movies.head()
```

# Preparing Keywords

```python
movies['keywords'] = movies['keywords'].apply(convert_str_to_list)
movies.head()
```

# Preparing Cast

```python
ast.literal_eval(movies['cast'][0])[0]
```

```python
def convert_cast(obj):
  new_list = []
  count = 0
  for i in ast.literal_eval(obj):
    new_list.append(i['name'])
    count+=1
    if count == 3:
      break
  
  return new_list
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
movies.sample(2)
```

# Preparing Crew

```python
ast.literal_eval(movies['crew'][0])
```

```python
def fetch_crew(obj):
  new_list = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      new_list.append(i['name'])
      break
    
  return new_list
```

```python
movies['crew'] = movies['crew'].apply(fetch_crew)
```

```python
movies.sample(3)
```

# Preparing Overview

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head(2)
```

# Removing spaces

```python
def remove_spaces(obj):
  new_list = []
  for i in obj:
    new_list.append(i.replace(' ',''))
  return new_list
```

```python
movies['genres'] = movies['genres'].apply(remove_spaces)
movies['keywords'] = movies['keywords'].apply(remove_spaces)
movies['cast'] = movies['cast'].apply(remove_spaces)
movies['crew'] = movies['crew'].apply(remove_spaces)
# movies.head(2)
```

# Creating tags column

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies['tags'][0]
```

```python
new_df = movies[['movie_id','title','tags']]
new_df.head(2)
```

# Creating a paragraph out of tags list

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df['tags'][0]
# new_df.head(2)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x : x.lower())
new_df['tags'][0]
```

# Text Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(new_df['tags']).toarray()
vectors
```

```python
cv.get_feature_names_out()
```

# Stemming

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
print(ps.stem('leaves'))
print(ps.stem('leaf'))
print(ps.stem('dance'))
print(ps.stem('dancing'))
```

```python
ps = PorterStemmer()
```

```python
def stemming(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stemming)
new_df['tags']
```

```python
new_df.head(2)
```

# Calculating similarity between the vectors

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
similarity
```

```python
list(enumerate(similarity[0]))
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x: x[1])
```

# Recommend Function

```python
def recommend(movie):
  movie_index = new_df[new_df['title'] ==  movie].index[0]
  distances = similarity[movie_index]
  distances = list(enumerate(distances))
  movies_list = sorted(distances, reverse=True, key=lambda x: x[1])[1:8]
  
  for i in movies_list:
    print(new_df.iloc[i[0]].title)

recommend('Batman')
  
```
